```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xs = np.arange(0, 5, 1)
alphap = 1
alphan = -1
ysd = np.exp(-alphap * xs)
ysu = np.exp(-alphan * xs)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))
axs[0].plot(xs, ysd, '-o')
axs[1].plot(xs, ysu, '-o')
for ax in axs:
    ax.set_xlim(0, 4)
    ax.set_xticks([0, 1, 2, 3, 4])
    ax.set_xticklabels([0, '$\Delta t$', '$2\Delta t$', '$3\Delta t$', '$4\Delta t$'])
    ax.set_xlabel('Time')
    ax.set_ylabel('y')
fig.savefig('lab1_graphs.pdf')
```
