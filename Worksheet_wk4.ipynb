### Question A ###

Compute the condition number for the matrix that arises from the Heat Equation using Dirchlet BC for various values of the number of segments, N, the rod is divided into. Note that N segments means N+1 grid points. Consider values of N between 5 and 50.  Remember that python starts counting from 0 (not 1).

(Hint: This will be much easier if you write a small Python function that outputs the matrix for a given value of N.)

The Matrix is:


$$A_1 = {\left[
          \begin{array}{ccccccccc}
            1&  0  & & \dots  & & & & & 0 \\
            1& {-2} & {1} & {0} & {\dots} & && &  \\
            0& {1} & {-2} &  {1} & {0} & {\dots} & & &  \\
            & {0} & {1}  & {-2} & {1} & {0} & {\dots} & &  \\
            & & & & & & & & \\
            \vdots & & & {\ddots} & {\ddots} & {\ddots} & {\ddots} & {\ddots} & \vdots \\
            & & & & & & & & \\
            & & & {\dots} & {0} & {1} & {-2} & {1} & 0 \\
            & &  & &{\dots} & {0} & {1} & {-2} & 1 \\
            0& & & & & \dots & & 0 & 1 \\
          \end{array}
        \right]}
$$    

```python
# function that outputs matrix for a given value of N

def matrix_N(n):
    matrix = np.zeros((n,n))
    matrix[-1][-1]=1
    matrix[0][0]=1
    for i in range(n-2):
        matrix[i+1][i]=1
        matrix[i+1][i+1]=-2
        matrix[i+1][i+2]=1
    return matrix
```

```python
# cell to calculate the condition number for various N
cond_numbers = []
for i in range(46):
    print("i is " + str(i) + " and the cond number is " + str(np.linalg.cond(matrix_N(i+5))))
    cond_numbers.append(np.linalg.cond(matrix_N(i+5)))
```

### Question B ###

Plot your results on a log-log plot (that is log condition number versus log N)

Also plot N$^2$ on the same plot.

How does the conition number of A$_1$ depend on N?

```python
# code to do the plot

N = np.arange(5,51,1)

plt.loglog(N, cond_numbers, label='cond vs N')
plt.xlabel("Matrix size")
plt.ylabel("Condition Number")
plt.title("Cond number vs matrix size")
plt.loglog(N, N**2, label= 'N vs N**2')
plt.grid()
plt.legend()
```

The log of the condition number increases linearly with the log of N. The condition number is nearly quadratic, since we are comparing to a quadratic equation. If the lines were parallel, then we would know that the condition number increases quadratically.

$$A_2 = {\left[

Does this change in the matrix make a significant difference in the

```python
# new function that outputs the matrix A_2 for a given value of N

def matrix_N_2(n):
    matrix = np.zeros((n,n))
    matrix[-1][-1]=-2
    matrix[-1][-2]=1
    matrix[0][0]=-2
    matrix[0][1]=1
    for i in range(n-2):
        matrix[i+1][i]=1
        matrix[i+1][i+1]=-2
        matrix[i+1][i+2]=1
    return matrix
```

```python
# cell to calculate the condition number for various N
cond_numbers_2 = []
for i in range(46):
    print("i is " + str(i) + " and the cond number is " + str(np.linalg.cond(matrix_N_2(i+5))))
    cond_numbers_2.append(np.linalg.cond(matrix_N_2(i+5)))
```

```python
# log log plot comparing
plt.loglog(N, cond_numbers_2, label='N vs cond')
plt.xlabel("Matrix size")
plt.ylabel("Condition Number")
plt.title("Cond number vs matrix size")
plt.loglog(N, N**2, label='N vs N**2')
plt.grid()
plt.legend()
```
